```python
import osmnx as ox
import matplotlib.pyplot as plt
import pandas as pd
import mcount
import numpy as np
```

# Some Results

```python
raw_data = pd.read_csv("cities_with_population.csv")
data = raw_data[raw_data.Nodes.notnull()]
data = data[data.Population.notnull()]
data = data.drop(columns =['Unnamed: 0', 'Unnamed: 0.1'])
data
```

```python
for column in list(data.columns.values)[6:15]:
    ax = data.plot(kind='scatter',x='Essential edges',y=column,color='red')
    fig = ax.get_figure()
    fig.savefig("figures/scatter_edges/" + str(column) + '_edges.png', bbox_inches='tight')
```

Now let's try with the number of non-nested motifs instead. I will create a new dataframe for that:

```python
#First we copy the subgraph chunk of the dataframe. Note that I don't count the self-loops here, as they are non-nested:
data_sg = data[list(data.columns.values)[7:15]]
sg_arr = data_sg.values
sg_nnest_arr = []

for row in sg_arr:
    motifs = row
    nnest = mcount.get_nnest_motifvector(motifs)
    sg_nnest_arr.append(nnest)

sg_nnest_arr = np.asarray(sg_nnest_arr)
#print(sg_nnest_arr.shape)
#print(sg_nnest_arr)
```

```python
data_nnest_sg = pd.DataFrame(data = sg_nnest_arr,columns=list(data.columns.values)[7:15])
```

```python
#Now we need to concatenate all pieces of our dataframe into a data_2 df:
df_1 = data[list(data.columns.values)[:7]]
df_2 = data_nnest_sg
df_3 = data["Population"]

#reset index: 
df_1 = df_1.reset_index(drop=True)
df_2 = df_2.reset_index(drop=True)
df_3 = df_3.reset_index(drop=True)

#concatenate all 3:
data_nonnested = pd.concat([df_1, df_2, df_3], axis=1)
```

```python
data_nonnested
```

```python
for column in list(data_nonnested.columns.values)[6:15]:
    data_nonnested.plot(kind='scatter',x='Population',y=column,color='red')
```

Now I will create a dataframe with the expected number of each motif in a random graph. At the same time, I create a dataframe with the nonnested version of that:

```python
#For the expected value, we need the count of nodes and edges. I use the essential edges, i.e. after simplification.
data_nodes = data[['Nodes', 'Essential edges']]
parameters_arr = data_nodes.values

random_arr = []
random_nnest_arr = []

for row in parameters_arr:
    #Each of these rows is a list of two parameters: the number of nodes and the number of edges.
    n = row[0]
    m = row[1]
    
    random_vector = mcount.get_random_motifvector(n, m)
    random_nnest_vector = mcount.get_randomnnest_motifvector(random_vector, n, m)
    
    random_arr.append(random_vector)    
    random_nnest_arr.append(random_nnest_vector)

#Now we just turn that list into a dataframe and concatenate the dataframes like before:
random_arr = np.asarray(random_arr)
random_nnest_arr = np.asarray(random_nnest_arr)

data_random_sg = pd.DataFrame(data = random_arr, columns=list(data.columns.values)[7:15])
data_random_nnest_sg = pd.DataFrame(data = random_nnest_arr, columns=list(data.columns.values)[7:15])

df_1 = data[list(data.columns.values)[:7]]
df_1 = df_1.reset_index(drop=True)
df_2 = data_random_sg
df_2 = df_2.reset_index(drop=True)
df_3 = data["Population"]
df_3 = df_3.reset_index(drop=True)
data_random = pd.concat([df_1, df_2, df_3], axis=1)

df_2 = data_random_nnest_sg
df_2 = df_2.reset_index(drop=True)
data_nnest_random = pd.concat([df_1, df_2, df_3], axis=1)
```

```python
for column in list(data_nnest_random.columns.values)[6:15]:
    data_nonnested.plot(kind='scatter',x='Population',y=column,color='red')
```

I will now calculate in an array the abundance of each tetradic motif:

```python
#First we use the chunks of dataframes we had before and adjust their indices:
data_tetrads = data_sg[list(data_sg.columns.values)[2:]].reset_index(drop=True)
data_random_tetrads = data_random_sg[list(data_random_sg.columns.values)[2:]].reset_index(drop=True)

#Now we define the formula for the abundance. Notice that we need a tolerance, which I borrow from Milo to be tol=4 for tetrads
tol = 4
num_df = data_tetrads - data_random_tetrads
den_df = data_tetrads + data_random_tetrads + tol
data_tetrad_abd = num_df.div(den_df)

#To present the significance profile, we will need it to be normalized:
sp4_arr = []
for row in data_tetrad_abd.values:
    norm = np.linalg.norm(row)
    sp4 = row/norm
    sp4_arr.append(sp4)
sp4_arr = np.asarray(sp4_arr)
data_tetrad_sp = pd.DataFrame(data = sp4_arr,columns=list(data_tetrad_abd.columns.values))

#And we add the population and names of cities:
df_1 = data[list(data.columns.values)[:3]]
df_1 = df_1.reset_index(drop=True)
df_2 = data_tetrad_sp
df_2 = df_2.reset_index(drop=True)
df_3 = data["Population"]
df_3 = df_3.reset_index(drop=True)
data_tetrad_sp_full = pd.concat([df_1, df_2, df_3], axis=1)

#Now I plot it
x = [1, 2, 3, 4, 5, 6]
for row in data_tetrad_sp.values:
    y = row
    plt.plot(x, y)

plt.xlabel('Tetradic motif')
plt.ylabel('Relative subgraph ocurrences')

plt.axhline(color='black', lw=1)
ax = plt.axes()
ax.xaxis.grid(True)

fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

ticks = x
labels = ["Path", "Complete", "Star", "Square", "Diamond", "Tadpole"]
plt.xticks(ticks, labels)

plt.show()

fig.savefig("figures/sp_tetrad/sp_all.png", bbox_inches='tight')
```

Repeat, but calculating abundance with relation to non-nested motifs now:

```python
#First we use the chunks of dataframes we had before and adjust their indices:
data_nnest_tetrads = data_nnest_sg[list(data_nnest_sg.columns.values)[2:]].reset_index(drop=True)
data_nnest_random_tetrads = data_random_nnest_sg[list(data_random_nnest_sg.columns.values)[2:]].reset_index(drop=True)

#Now we define the formula for the abundance. Notice that we need a tolerance, which I borrow from Milo to be tol=4 for tetrads
tol = 4
num_nnest_df = data_nnest_tetrads - data_nnest_random_tetrads
den_nnest_df = data_nnest_tetrads + data_nnest_random_tetrads + tol
data_nnest_tetrad_abd = num_nnest_df.div(den_nnest_df)

#To present the significance profile, we will need it to be normalized:
sp4_nnest_arr = []
for row in data_nnest_tetrad_abd.values:
    norm = np.linalg.norm(row)
    sp4 = row/norm
    sp4_nnest_arr.append(sp4)
sp4_nnest_arr = np.asarray(sp4_nnest_arr)
data_nnest_tetrad_sp = pd.DataFrame(data = sp4_nnest_arr,columns=list(data_nnest_tetrad_abd.columns.values))

#And we add the population and names of cities:
df_1 = data[list(data.columns.values)[:3]]
df_1 = df_1.reset_index(drop=True)
df_2 = data_nnest_tetrad_sp
df_2 = df_2.reset_index(drop=True)
df_3 = data["Population"]
df_3 = df_3.reset_index(drop=True)
data_nnest_tetrad_sp_full = pd.concat([df_1, df_2, df_3], axis=1)

#Now I plot it
x = [2,6,1,4,5,3]
for row in data_nnest_tetrad_sp.values:
    y = row
    plt.scatter(x, y)

plt.xlabel('Tetradic motif')
plt.ylabel('Relative subgraph ocurrences')

plt.axhline(color='black', lw=1)
ax = plt.axes()
ax.xaxis.grid(True)

fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

ticks = x
labels = ["Path", "Complete", "Star", "Square", "Diamond", "Tadpole"]
plt.xticks(ticks, labels)

plt.show()

fig.savefig("figures/sp_tetrad/ticks_ordered.png", bbox_inches='tight')
```
